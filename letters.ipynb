```python
from tabulate import tabulate
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
```

```python
def fileToSequence():
    sample_text = 'testing/data/text/finnish.txt'
    with open(sample_text) as sample_text:
        sample = sample_text.read()
    
    i = text_to_word_sequence(
        sample,
        filters='!"\'#$%&()*+,-./:;<=>?@[\]^_,\n',
        lower=True, 
        split=' '
    )
    return i
wordSequence = fileToSequence()
```

```python
num_words = 1000
oov_token = '<UNK>'
pad_type = 'post'
trunc_type = 'post'
tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)
tokenizer.fit_on_texts(wordSequence)
```

```python
letters = 'aeiouäö'
words = wordSequence
```

```python
def letterEnvironment(letters):
    letters_words = {}
    def letterCheck(letter):
        wordEnvironment = []
        for w in words:
            modifiedWord = w
            lettersMatched = 0
            while True:
                letterIndex = modifiedWord.find(letter)
                if letterIndex == -1:
                    break
                else:
                    lowerBound = letterIndex + lettersMatched - 1
                    upperBound = letterIndex + lettersMatched + 2
                    if lowerBound < 0:
                        lowerBound = 0
                    wordEnvironment.append(w[lowerBound:upperBound])
                    lettersMatched = lettersMatched + 1
                    modifiedWord = modifiedWord[:letterIndex] + modifiedWord[letterIndex+1:]
        d1 = {letter: wordEnvironment}
        letters_words.update(d1)
    letters = [l for l in letters]
    for n_letter in letters:
        letterCheck(n_letter)
    return letters_words
```

```python
x = letterEnvironment(letters)
print(tabulate(x, headers='keys', tablefmt='github'))
```
