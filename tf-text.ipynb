```python
import matplotlib.pyplot as plt
import os, string
import numpy as np
from tabulate import tabulate
from collections import OrderedDict, Counter
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence, hashing_trick
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def getLetters():
    letters_file = 'data/letters/letters.txt'
    with open(letters_file) as letter_file:
        letters = letter_file.read()
    return letters
```

```python
def fileToSequence():
    sample_text = 'data/text/finnish.txt'
    with open(sample_text) as sample_text:
        sample = sample_text.read()
    
    i = text_to_word_sequence(
        sample,
        filters='!"\'#$%&()*+,-./:;<=>?@[\]^_,\n',
        lower=True, 
        split=' '
    )
    return i
wordSequence = fileToSequence()
```

```python
num_words = 1000
oov_token = '<UNK>'
pad_type = 'post'
trunc_type = 'post'
tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)
tokenizer.fit_on_texts(wordSequence)
```

```python
letters = getLetters()
words = set(wordSequence)
```

```python
def letterEnvironment(letterList):
    letterEnvironment = []
    def letterCheck(letter):
        wordEnvironment = []
        letterEnvironment.append([letter])
        for w in words:
            modifiedWord = w
            lettersMatched = 0
            while True:
                letterIndex = modifiedWord.find(letter)
                if letterIndex == -1:
                    break
                else:
                    lowerBound = letterIndex + lettersMatched - 1
                    upperBound = letterIndex + lettersMatched + 2
                    if lowerBound < 0:
                        lowerBound = 0
                    wordEnvironment.append([w[lowerBound:upperBound]])
                    lettersMatched = lettersMatched + 1
                    modifiedWord = modifiedWord[:letterIndex] + modifiedWord[letterIndex+1:]
        letterEnvironment.append(wordEnvironment)
    letter = [key for key in letterList]
    for l in letters:
        letterCheck(l)
    return letterEnvironment

letterEnvironment(letters)
```
